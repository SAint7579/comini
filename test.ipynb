```python
import pandas as pd
```

```python
df = pd.read_csv('Data/tools_master.csv')
for col in df.columns:
    num_duplicates = df.duplicated(subset=[col], keep=False).sum()
    if num_duplicates > 0:
        print(f"Number of duplicates in column '{col}': {num_duplicates}")
        print('-' * 40)
```
